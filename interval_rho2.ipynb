This experiment evaluates the evolution of the error as the number of graphs to be inferred increases. It compares the performance of independently identifying K similar graphs with respect the performance of a joint inference algorithm. 

```python
import numpy as np
import matplotlib.pyplot as plt


SEED = 0
np.random.seed(SEED)
```

```python
alpha1 = np.linspace(0, 1, 200)
alpha2 = np.linspace(0, 1, 200)

Alpha1, Alpha2 = np.meshgrid(alpha1, alpha2)
# Reorder Alpha2 soy it grows with yaxis 
Alpha2 = Alpha2[::-1,:]

upper_bound = (2+Alpha1)*Alpha1/(4*(1-Alpha1))
lower_bound = (2-Alpha2)/Alpha2*(1+Alpha1/4)
valid_rho2 = upper_bound > lower_bound

plt.figure()
plt.imshow(lower_bound, extent=[0, 1, 0, 1])
plt.colorbar()
plt.xlabel('alpha1')
plt.ylabel('alpha2')
plt.title('Lower bound')

plt.figure()
plt.imshow(upper_bound, extent=[0, 1, 0, 1])
plt.colorbar()
plt.xlabel('alpha1')
plt.ylabel('alpha2')
plt.title('Upper bound')

plt.figure()
plt.imshow(valid_rho2, extent=[0, 1, 0, 1])
plt.colorbar()
plt.xlabel('alpha1')
plt.ylabel('alpha2')
plt.title('Valid rho2')

print('Proportion of valid rho2:', np.sum(valid_rho2)/valid_rho2.size)
```
